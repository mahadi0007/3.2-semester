```python
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt
from PIL import Image

```

```python
# Authenticate with Hugging Face
from huggingface_hub import login
login()


```

```python
# Load the Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# List of custom prompts
prompts = [
    "A futuristic city floating in the clouds, digital art",
    "A magical forest with glowing plants and fairies, fantasy style",
    "A hyper-realistic portrait of a cat astronaut in space",
    "A serene landscape with waterfalls and floating islands, Studio Ghibli style",
    "A 16th century knight fighting a dragon, oil painting"
]

```

```python
# Generate and show images for each prompt
def generate_images(prompts):
    for i, prompt in enumerate(prompts):
        image = pipe(prompt).images[0]
        plt.figure(figsize=(6,6))
        plt.imshow(image)
        plt.axis("off")
        plt.title(f"Prompt {i+1}: {prompt}")
        plt.show()

# Run generation
generate_images(prompts)
```
